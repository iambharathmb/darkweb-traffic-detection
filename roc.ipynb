# RandomForestClassifier

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import roc_curve, auc
import sklearn.utils.multiclass

# Load the Darknet dataset (assuming it is in CSV format)
data = pd.read_csv("model_samples.csv")

# Split data into (Train & Test)
x = data.drop(columns=["Label"])
y = data["Label"]

# Encode categorical features using one-hot encoding
x = pd.get_dummies(x)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Model Selection
from sklearn.ensemble import RandomForestClassifier

# Initialize and Model
clf = RandomForestClassifier()

# Train the Model
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluation of the model
# Accuracy Score
acc = accuracy_score(y_test, y_pred)
print("RandomForestClassifier-Accuracy : ", acc)

# Confusion Matrix Display
cm = confusion_matrix(y_test, y_pred, labels= clf.classes_)
color = 'white'
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
ax = disp.plot()
disp.ax_.set_title("Confusion Matrix - RandomForestClassifier")
plt.show()

# Classification Report (Precision, Recall, F1 Score)
cr = classification_report(y_test, y_pred)
print("Classification Report : \n", cr)

# ROC Curve Display
def convert_output(y_pred):
  output = []
  for prediction in y_pred:
    if prediction == "Benign":
      output.append(0)
    else:
      output.append(1)
  return output
y_pred = convert_output(y_pred)

def convert_output(y_test):
  output = []
  for prediction in y_test:
    if prediction == "Benign":
      output.append(0)
    else:
      output.append(1)
  return output
y_test = convert_output(y_test)

fpr, tpr, thresholds = sklearn.metrics.roc_curve(y_test, y_pred, pos_label=1)
roc_auc = auc(fpr, tpr)

print("ROC Curve Acc - RandomForestClassifier : ")

# Plot the ROC curve
plt.plot(fpr, tpr, label="ROC curve (area = {0:.2f})".format(roc_auc))
plt.title("ROC Curve - RandomForestClassifier")
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.legend()
plt.show()
```
